```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
df = pd.read_csv('Spotify-2000.csv')
```

```python
# Fungsi matriks korelasi
def plotCorrelationMatrix(df, graphWidth):
    df = df.dropna('columns') # drop columns with NaN
    df = df[[col for col in df if df[col].nunique() > 1]] # keep columns where there are more than 1 unique values
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({df.shape[1]}) is less than 2')
        return
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.title(f'Matriks Korelasi', fontsize=15)
    plt.show()
```

# Tabel korelasi

```python
# Menunjukkan korelasi semua atribut kuantitatif dalam tabel
dfc = df.corr()
display(dfc)
```

Nilai korelasi yang mendekati 1 menandakan korelasi yang lurus, nilai korelasi yang mendekati -1 menandakan korelasi yang terbalik, dan nilai korelasi yang mendekati nol menandakan tidak ada korelasi antara kedua data tersebut.

Dapat dilihat bahwa angka korelasi yang bermakna ada di antara Loudness dengan Energy, Acousticness dengan Energy, Valence dengan Danceability, dan Acousticness dengan Loudness.

Maknanya akan dijelaskan di bagian selanjutnya.

# Matriks korelasi

```python
# Menunjukkan korelasi semua atribut kuantitatif dalam matriks korelasi berwarna
plotCorrelationMatrix(df, 8)
```

Dari data di atas dapat dilihat data-data yang memiliki korelasi dengan makna yang cukup jelas adalah:

1. Loudness dan Energy berkorelasi lurus
2. Acousticness dan Energy berkorelasi terbalik
3. Valence dan Danceability berkorelasi lurus
4. Acousticness dan Loudness berkorelasi terbalik

Data-data lain kurang berkorelasi.
