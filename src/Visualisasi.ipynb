```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

df = pd.read_csv('Spotify-2000.csv', thousands=r',')
```

# Perbandingan kategori

```python
# Perbandingan kategori
# Membandingkan popularitas 15 genre teratas
# Popularitas dihitung berdasarkan mean
df2 = df.groupby('Top Genre').mean()[['Popularity']]
df2 = df2.sort_values(['Popularity'], ascending = [0]).head(15)
df2.plot(kind="bar", figsize=(15, 15), width=0.6, color='#2ecc71')
plt.yticks(np.arange(0, 90, 5))
plt.title('Perbandingan Popularitas 15 Top Genre Teratas 1956-2019')
plt.ylabel('Rata-rata Popularitas 1956-2019')
plt.show()
```

Dari grafik dapat dilihat bahwa ada dua genre yang rata-rata popularitasnya paling tinggi, yaitu celtic punk dan indie pop.

```python
# Perbandingan kategori
# Membandingkan BPM 15 genre terendah
# BPM dihitung berdasarkan mean
df3 = df.groupby('Top Genre').mean()[['Beats Per Minute (BPM)']]
df3 = df3.sort_values(['Beats Per Minute (BPM)'], ascending = [1]).head(15)
df3.plot(kind="bar", figsize=(15, 15), width=0.6, color='#3498db')
plt.yticks(np.arange(0, 90, 5))
plt.title('Perbandingan Beats per Minute (BPM) 15 Top Genre Teratas 1956-2019')
plt.ylabel('Rata-rata Beats per Minute (BPM) 1956-2019')
plt.show()
```

Dari grafik dapat dilihat bahwa ada genre dengan rata-rata BPM paling lambat adalah indie anthem-folk.

# Penampilan perubahan terhadap waktu

```python
# Penampilan perubahan terhadap waktu
# Menampilkan popularitas album rock dari tahun 1977-2019
df4 = df.loc[df['Top Genre'] == 'album rock']
df4 = df4.groupby('Year')['Popularity'].mean()
df4.plot(kind='line', x='Year', y='Popularity', figsize=(12, 12), color='#e67e22')
plt.xticks(np.arange(1963, 2020, 4))
plt.yticks(np.arange(20, 85, 5))
plt.xlabel('Tahun')
plt.ylabel('Skala Popularitas')
plt.title('Popularitas Album Rock dari Tahun 1963-2019')
plt.legend(['Popularitas'])
plt.show()
```

Dari grafik, dapat dilihat perkembangan popularitas album rock dari tahun ke tahun. Dari tahun ke tahun, popularitasnya cenderung menurun dengan tingkat popularitas tertingginya berada di antara 1995-1999.

```python
# Penampilan perubahan terhadap waktu
# Menampilkan perubahan Danceability, Acousticness, dan Liveness rata-rata lagu 
# dari tahun 1956-2019

tdf1 = df.groupby('Year')['Danceability'].mean()
tdf1.plot(kind='line', figsize=(15, 15), color='#2ecc71')
tdf2 = df.groupby('Year')['Acousticness'].mean()
tdf2.plot(kind='line', color='#3498db')
tdf3 = df.groupby('Year')['Liveness'].mean()
tdf3.plot(kind='line', color='#9b59b6')
plt.legend(['Danceability', 'Acousticness', 'Liveness'])
plt.title('Rata-Rata Danceability, Acousticness, dan Liveness Lagu dari 1956-2019')
plt.xlabel('Tahun')
plt.ylabel('Skala 0-100')
plt.yticks(np.arange(0, 100, 5))
plt.xticks(np.arange(1956, 2020, 3))
plt.show()
```

Dari grafik, dapat dilihat perkembangan rata-rata Danceability, Acousticness, dan Liveness lagu-lagu per tahun.
Acousticness cenderung menurun dari tahun ke tahun, sedangkan danceability dan liveness cenderung stabil.

# Penampilan hierarki dan hubungan keseluruhan-bagian

```python
# Penampilan hierarki dan hubungan keseluruhan-bagian
# Persentase banyak munculnya data pada data ini per genre.
# Hanya ditampilkan 10 genre dengan frekuensi terbanyak.
pdf = df['Top Genre'].value_counts()[0:10]
pdf.plot(kind='pie', figsize=(12,12), autopct='%1.1f%%', shadow=True)
plt.ylabel('')
plt.title('Pie Chart Frekuensi Munculnya Genre Lagu pada Dataset Keseluruhan')
plt.show()
```

Dari pie chart di atas dapat dilihat bahwa genre album rock mendominasi dataset ini apabila meninjau semua lagu yang di dataset dari 1956-2019.

```python
# Penampilan hierarki dan hubungan keseuluruhan-bagian
# Persentase banyak munculnya data pada data ini per genre.
# Hanya ditampilkan 10 genre dengan frekuensi terbanyak.
pdf2 = df.loc[df['Year'] == 2019]
pdf2 = pdf2['Top Genre'].value_counts()
pdf2.plot(kind='pie', figsize=(12,12), autopct='%1.1f%%', shadow=True)
plt.ylabel('')
plt.title('Pie Chart Frekuensi Munculnya Genre Lagu pada Dataset di Tahun 2019')
plt.show()
```

Berbeda dengan kesimpulan sebelumnya, apabila meninjau lagu-lagu terbitan tahun 2019, genre album rock mampu disaingi oleh dutch pop.

# Plotting relationships

```python
# Plotting relationships
# Hubungan antara Acousticness dan Energy
df.plot(kind='scatter', x='Acousticness', y='Energy', color='#e74c3c')
plt.title('Scatter plot antara Acousticness dan Energy')
corr = df['Acousticness'].corr(df['Energy'])
plt.legend([f'Nilai korelasi: {corr}'])
plt.show()
```

Dari gambar di atas, walaupun tidak terlalu jelas (nilai mutlak korelasi ~0,66), ditunjukkan suatu korelasi negatif antara acousticness dan energy. Semakin tinggi acousticness, cenderung semakin turun nilai energy.

```python
# Plotting relationships
# Hubungan antara Acousticness dan Energy
df.plot(kind='scatter', x='Loudness (dB)', y='Energy', color='#d35400')
plt.title('Scatter plot antara Loudness dan Energy')
corr = df['Loudness (dB)'].corr(df['Energy'])
plt.legend([f'Nilai korelasi: {corr}'])
plt.show()
```

Dari gambar di atas, walaupun tidak terlalu jelas (nilai mutlak korelasi ~0,74), ditunjukkan suatu korelasi positif antara loudness dan energy. Semakin tinggi loudness, cenderung semakin tinggi nilai energy.
