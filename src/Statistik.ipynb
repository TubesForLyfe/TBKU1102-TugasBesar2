# Statistik rata-rata dan standar deviasi

```python
# Statistik rata-rata data dan standar deviasi.
print('Rata-rata:')
display(df[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence', 'Length (Duration)',
   'Acousticness', 'Speechiness', 'Popularity']].mean())
print('Standar deviasi:')
display(df[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence', 'Length (Duration)',
   'Acousticness', 'Speechiness', 'Popularity']].std())
```

# Statistik percentile

```python
# Statistik percentile (10%, 25%, 50%, 75%, dan 90%).
print('Persentil:')
display(df[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence', 'Length (Duration)',
   'Acousticness', 'Speechiness', 'Popularity']].quantile([0.10, 0.25, 0.50, 0.75, 0.90]))
```

# Statistik ekstremum (nilai maksimum dan minimum)

```python
# Statistik ekstremum maksimum dan minimum
print('Maksimum: ')
display(df[['Year', 'Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence', 'Length (Duration)',
   'Acousticness', 'Speechiness', 'Popularity']].max())
print('Minimum: ')
display(df[['Year', 'Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence', 'Length (Duration)',
   'Acousticness', 'Speechiness', 'Popularity']].min())
```

```python
# Statistik untuk setiap genre.
print('Mean:')
display(df.groupby('Top Genre')[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence',
   'Acousticness', 'Speechiness', 'Popularity']].mean().dropna())
print('Standar Deviasi:')
display(df.groupby('Top Genre')[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence',
   'Acousticness', 'Speechiness', 'Popularity']].std().dropna())
print('Nilai Maksimum:')
display(df.groupby('Top Genre')[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence',
   'Acousticness', 'Speechiness', 'Popularity']].max().dropna())
print('Nilai Minimum:')
display(df.groupby('Top Genre')[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence',
   'Acousticness', 'Speechiness', 'Popularity']].min().dropna())
```

# Statistik per genre lagu

```python
# Mencari informasi-informasi yang menarik per genre.
display(df.groupby('Top Genre')[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence',
   'Acousticness', 'Speechiness', 'Popularity']].mean().dropna().sort_values('Beats Per Minute (BPM)'))
print('Rata-rata menunjukkan genre indie anthem-folk memiliki BPM paling rendah dan genre laboratorio memiliki BPM paling tinggi.')
display(df.groupby('Top Genre')[['Beats Per Minute (BPM)', 'Energy', 
   'Danceability', 'Loudness (dB)', 'Liveness', 'Valence',
   'Acousticness', 'Speechiness', 'Popularity']].mean().dropna().sort_values('Popularity'))
print('Rata-rata menunjukkan genre classic italian pop memiliki Popularity paling rendah dan genre indie pop dan celtic punk memiliki Popularity paling tinggi.')
```
